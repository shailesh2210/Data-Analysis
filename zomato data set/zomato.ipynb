```python
df1 = pd.read_excel("Country-Code.xlsx")
```

```python
df1.head()
```

```python
# combine two data frames
data = pd.merge(df, df1 ,on="Country Code", how="left")
```

```python
data.head(3)
```

```python
data.Country.value_counts()
```

- zomato mostly available in India

```python
country_names = data.Country.value_counts().index
```

```python
country_value = data.Country.value_counts().values
```

```python
# pie chart

plt.pie(country_value , labels=country_names)
```

```python
# pie chart -- Top 3 countries that uses zomato
plt.pie(country_value[:3], labels=country_names[:3] ,autopct= "%1.1f%%" )
```

- Observation : Zomato maximum records are from india
- Ater USA and then United Kingdoms

```python
data.groupby(["Aggregate rating", "Rating color", "Rating text"])
```

```python
data.groupby(["Aggregate rating", "Rating color", "Rating text"]).size()
```

```python
rating = data.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
rating
```

### Observation
- when rating is between 4.5 to 4.9 ---> Excellent
- when rating is between 4.0 to 4.5 ---> Very Good
- when rating is between 3.5 to 3.9 ---> Good
- when rating is between 2.5 to 3.4 ---> Average
- when rating is between 1.8 to 2.4 ---> Poor

```python
plt.rcParams["figure.figsize"] = (12, 6)    # changing size
sns.barplot(x = "Aggregate rating" , y="Rating Count" , data = rating)
```

```python
plt.rcParams["figure.figsize"] = (12, 6)    # changing size
sns.barplot(x = "Aggregate rating" , y="Rating Count" , data = rating , hue= "Rating color")
```

```python
# mapping colors

plt.rcParams["figure.figsize"] = (12, 6)    # changing size
sns.barplot(x = "Aggregate rating" , y="Rating Count" , data = rating , hue= "Rating color", palette=["Blue", "red", "Orange","Yellow","Green", "green"])
```

### Observation:
1. Not rated count is very high
2. maximum number of rating are between 2.5 to 3.9

```python
# Count plot

sns.countplot(x ="Rating color", data = rating , palette=["Blue", "red", "Orange","Yellow","Green", "green"])
```
